```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
#read data as dataframes
#data generated by RF_XG.ipynb, run RF_XG.ipynb to create XC_RD.csv and XC_XG.csv
fp1 = '/Users/junyaokang/Desktop/TE/ML_Pyrite/XC_RD.csv'
fp2 = '/Users/junyaokang/Desktop/TE/ML_Pyrite/XC_XG.csv'
RD_Data = pd.read_csv(fp1)
XG_Data = pd.read_csv(fp2)
```

```python
#Split into Cui et al. and Xiao et al. data
RD_Data_Cui = RD_Data[RD_Data['Sample No'].isin(['Hy31', 'Hy59', 'Hy55', 'Hy1'])]
RD_Data_Xiao = RD_Data[RD_Data['Sample No'].isin(['JLW23.5', 'JLW15.1', 'JLW76', 'JLW34.8', 'SX88.15'])]
XG_Data_Cui = XG_Data[XG_Data['Sample No'].isin(['Hy31', 'Hy59', 'Hy55', 'Hy1'])]
XG_Data_Xiao = XG_Data[XG_Data['Sample No'].isin(['JLW23.5', 'JLW15.1', 'JLW76', 'JLW34.8', 'SX88.15'])]
```

```python
#labels: sedimentary (0), synsedimentary hydrothermal (1), and post-sedimentary hydrothermal(2)
fig = px.scatter_ternary(RD_Data_Cui, a='0', b='1', c='2', color='Sample No', symbol='Sample No', 
                        color_discrete_map={
                            'Hy31': 'gray',
                            'Hy59': 'steelblue',
                            'Hy55': 'lightskyblue',
                            'Hy1': 'peru'
                        },
                        opacity=0.6)
fig.update_traces(marker={'size': 10})
fig.show()
fig.write_image("Ternary_RD_Cui.svg")
```

```python
fig = px.scatter_ternary(RD_Data_Xiao, a='0', b='1', c='2', color='Sample No', symbol='Sample No',
                        color_discrete_map={
                            'JLW23.5': 'tan',
                            'JLW15.1': 'darkseagreen',
                            'JLW76': 'slategray',
                            'SX88.15': 'skyblue',
                            'JLW34.8': 'navy',
                        },
                        opacity=0.6)
fig.update_traces(marker={'size': 10})
fig.show()
fig.write_image("Ternary_RD_Xiao.svg")
```

```python
fig = px.scatter_ternary(XG_Data_Cui, a='0', b='1', c='2', color='Sample No', symbol='Sample No',
                        color_discrete_map={
                            'Hy31': 'gray',
                            'Hy59': 'steelblue',
                            'Hy55': 'lightskyblue',
                            'Hy1': 'peru'
                        },
                        opacity=0.6)
fig.update_traces(marker={'size': 10})
fig.show()
fig.write_image("Ternary_XG_Cui.svg")
```
